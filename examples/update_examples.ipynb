```python
import datetime
from pathlib import Path
import geopandas as gpd
from network_wrangler.roadway.io import load_roadway_from_dir, write_roadway
from network_wrangler.roadway.clip import clip_roadway
import logging

logging.basicConfig(level=logging.DEBUG)
```

```python
BASE_PATH = Path("/Users/elizabeth/Documents/urbanlabs/MetCouncil/NetworkWrangler/working")
assert BASE_PATH.is_dir()

BOUNDING_BOX = BASE_PATH / "network_wrangler/tests/data/unionstation.geojson"
assert BOUNDING_BOX.is_file()
NAME = "unionstation"

ROAD_PATH = BASE_PATH / "network/roadway"
assert ROAD_PATH.is_dir()
ROAD_OUT_PATH = ROAD_PATH / "clipped/unionstation/"
assert ROAD_OUT_PATH.is_dir()
ROAD_FMT = "geojson"

TRANSIT_PATH = BASE_PATH / "network/transit"
assert TRANSIT_PATH.is_dir()
TRANSIT_OUT_PATH = TRANSIT_PATH / "clipped/unionstation/"
assert TRANSIT_OUT_PATH.is_dir()
TRANSIT_FMT = "csv"
```

```python
from network_wrangler.utils.io import geojson_to_parquet
from network_wrangler.roadway.io import id_roadway_file_paths_in_dir

STPAUL_GEO = Path("./stpaul/downtown_stpaul_boundary.geojson")
stpaul_geo = gpd.read_file(STPAUL_GEO)
links_file, nodes_file, shapes_file = id_roadway_file_paths_in_dir(ROAD_PATH, "geojson")

geojson_to_parquet(nodes_file, ROAD_OUT_PATH / "stpaul_nodes.parquet", filter_to_geo=stpaul_geo)
```

```python
net = load_roadway_from_dir(ROAD_PATH)
clipped_net = clip_roadway(net, boundary_file=BOUNDING_BOX)
write_roadway(
    clipped_net,
    prefix=NAME,
    out_dir=ROAD_OUT_PATH,
    file_format=ROAD_FMT,
    overwrite=True,
)
readme_txt = f"{NAME}.{ROAD_FMT} Network was clipped using clip_roadway().\n\
    Date: {datetime.datetime.now().strftime('%Y_%m_%d__%H_%M_%S')}\n\
    Input: {ROAD_PATH}\n\
    Bounding Box: {BOUNDING_BOX}\n"
readme_path = ROAD_OUT_PATH / f"{NAME}_{ROAD_FMT}_README.txt"
with open(readme_path, "w") as f:
    f.write(readme_txt)
```
